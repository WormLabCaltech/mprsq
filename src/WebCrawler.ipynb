```python
import requests
from bs4 import BeautifulSoup
import os 

# website containing all the samples
# web= 'https://jumpgate.caltech.edu/\
# runfolders/volvox02/flowcells/HT2GVBCXX/C1-116/'
web= 'https://jumpgate.caltech.edu/runfolders/volvox02/160808_SN787_0534_AHVGMLBCXX/Unaligned/'

# open the website
resp = requests.get(web, auth=('gec', 'gecilluminadata'))

# encode it
encoding = resp.encoding if 'charset' in resp.headers.get('content-type', '').lower() else None

# beautiful soup to extract html links:
soup = BeautifulSoup(resp.content, "lxml", from_encoding=encoding)

# get every project name so that we can get the URLS. 
projects = []
for link in soup.find_all('a', href=True):
    if'Project' in link['href']:
        projects += [link['href']]


path = '../input/rawseq/'
for project in projects:
    w = web+project
    resp = requests.get(w, auth=('gec', 'gecilluminadata'))
    
    # beautiful soup to extract html links:
    soup = BeautifulSoup(resp.content, "lxml", from_encoding=encoding)
    
    # there's a link called sample, get it:
    link = soup.find_all('a', href=True)
    sample = link[5]['href']  # fifth item down is href
    
    # open the new website:
    w2 = web + project + sample
    resp = requests.get(w2, auth=('gec', 'gecilluminadata'))
    soup = BeautifulSoup(resp.content, "lxml", from_encoding=encoding)
    i = 0
    os.mkdir(path+project)
    
    for link in soup.find_all('a', href=True):
        file = link['href']
        print(file)
        if i > 4:
            download = web + project + sample + file
            r = requests.get(download, auth=('gec', 'gecilluminadata'))
            with open(path+project+file, "wb") as code:
                code.write(r.content)
        i+=1
```
